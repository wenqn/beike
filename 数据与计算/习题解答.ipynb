
### 【练习二】用python完成下列计算
---

-  给定一个正整数，取出指定的任意位置上的数字。比如数字1786，要取出数字7，可以按如下步骤
  > |步骤|思路|操作|实现|结果|
  >|--|--|--|--|--|
  >|1|将7后面的数字截掉，让7作为个位数，即1786把7后面的86去掉，变成17|1786除以100取整| 1786 // 100|17|
  >|2|将上一步的数字对10取模|对17用10进行模运算|17%10|7|

  现在有一个数字123456，分别把数字2和数字6取出来.
  >  想一想：如何从整数123456中取出尾数‘56’？
  
- 网上购物平台搜索商品时可以通过条件进行筛选，以提高搜索的针对性。某同学要买一个手机支架，请用表达式表示出符合“包邮","赠运费险", “价格在30元-50元之间"，发货地在"江浙沪"和"山东"的商品。



- 取数字是个典型的数字游戏，主要用到整除和取模运算。特别是取模运算的思路需要重点了解。取模运算是求解整除后的余数（不是小数），经常用于求解咋周期内数字的排位
  - 数字123456，要取出数字2，2在万位上，先用整除把万位以下的数字截掉，123456//1000，得到12，再对10取模得到个位数2，12%10  
  - 数字6已经在个位上，直接对10取模即可，123456 % 10 
  - 取模运算的规则就是整除后的余数。取尾数‘56’，可以借鉴取个位数思路，对10取模得到个位数，对100取模得到个位和十位数，对1000取模得到个位十位和百位数。
- 这些条件要同时满足，要用逻辑“与”运算，要考虑最后的发货地，同一件商品不可能从两个地方同时发出来，所以，发货地是个“或”运算。可以借助数学上的集合概念来理解与、或、非的计算结果
  - 逻辑表达式："包邮" and "赠运费险" and (30 <= price <= 50>) and ((addr == "江浙沪") or (addr == "山东"))
  - 用变量来表示如下

```python
# 以下内容演示思路和结果即可，不必要求学生写出来

# 定义变量 
free_delivery = True   # 包邮
free_freight_insurance = True  #赠运费险
price = 45  # 定义价格
addrs = ["江浙沪","江苏","浙江","上海","山东"]  # 定义可接受的发货地址范围
addr = "山东"  # 定义实际发货地址

# 表达式，运行结果为True即符合题意
free_delivery and free_freight_insurance and  (30 <= price <= 50) and (addr in addrs) 
```

> `in` 是成员符号 ，表达式 `a in b` 的意思是判断变量a是否包含在b中，因此b一般是 字符串、列表、元组或字典等数据类型。
