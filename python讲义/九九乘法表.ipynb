用for循环打印九九乘法表。

### IPO分析
- I：无输入
- P：计算中间结果。每行每列都不同，每一行比上一行多一列。需要用两个变量分别控制，i控制行，j控制列。需要用两层循环来实现。
- O：打印出三角形的九九乘法表，每一组用`j*i=结果`形式打印，每一行完成一组乘数i的计算。

### 循环嵌套
for 嵌套就是：for 里面还有 for

```
for i in 循环条件 1:
    条件满足时，做的事情1
    条件满足时，做的事情2
    条件满足时，做的事情3
    ...(省略)...

    for i in 循环条件 2:
        条件满足时，做的事情1
        条件满足时，做的事情2
        条件满足时，做的事情3
        ...(省略)...
```

````python
#  九九乘法表
#定义结果和中间变量
result=0
temp=0
#九九乘法表中一共有九行，但每行列数不同所以通过i进行定义
for i in range(1,10):

    # j 控制每一行的列数
    for j in range(i):
        temp=(j+1)  # range() 第一个值是0，要加1，从1开始计算
        result=temp*i
        print(temp,"*",i,"=",result,end="\t")

    print("\n")
````
