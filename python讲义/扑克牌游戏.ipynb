### 洗牌发牌
如何公平的洗牌是一个比较有深度的话题，经常成为IT公司的面试题目。所谓公平是指抽到的每张牌的概率是相等的，用数学上的话说就是抽到每张牌的概率是真正随机的。随机数在游戏里应用非常广泛，python里有专门处理随机数的功能包`random`。`random`自带一个洗牌函数`shutffle()`,可以实现基本公平的洗牌。使用random包之前要先导入该包。
#### 洗牌
```
import random           #导入random
random.shuttfle(cards)     #洗牌
```
发牌或模牌的过程就是从现有列表里随即抽取一张，要用到`random`里的`choice()`函数。摸完一张牌，要记得从`cards`里删除这张牌。可以用列表自带函数`remove()`函数。这个游戏中假定两个玩家，分别用`persionA`，`persionB`表示两人摸到的牌。
#### 发牌
```
#有问题的代码
persionA = random.choice(cards)
persionB = random.choice(cards)
```
上面的发牌程序是有问题的。你想一想问题在哪？
```
#正常的代码
persionA = random.choice(cards)
cards.remove(persionA)
persionB = random.choice(cards)
cards.remove(persionB)
```
现在还面临两个问题，一是变量`persionA`,`persionB`只能存储1张牌，要摸一把牌就不能用单个变量，而应该用列表。二是每摸一张牌就需要choice一次，再remove一次。一副牌摸下来要反反复复多次。可以用循环来简化代码，其实更好的方法是把摸牌的过程写成一个函数。
#### 发牌函数 dealcards(cards)
```
# cards:扑克牌列表
# a:返回抽出的牌
def dealcards(cards):
    if cards:   #如果cards不为空，也就是还没摸完牌
        a=random.choice(cards)
        cards.remove(a)
        random.shuttfle(cards)
        return a
```

#### 调用dealcards()函数
```
# persionA，persionB两个玩家轮流抽牌，直至摸完所有牌
persionA = []  #存储玩家A摸到的牌
persionB = []  #存储玩家B摸到的牌
while cards:
    persionA.append(dealcards(cards))
    persionB.append(dealcards(cards))
```
`dealcards()`函数一次只能摸一张牌，用条件循环`while`来实现连续摸牌。请考虑下下面的`while`循环与上面的`while`循环有何不同？
```
persionA = []
persionB = []
while cards:
    persionA.append(dealcards(cards)
    if cards:
        persionB.append(dealcards(cards))
```

````python
# 完整摸牌函数
def dealcards(cards):
    if cards:   #如果cards不为空，也就是还没摸完牌
        a=random.choice(cards)
        cards.remove(a)
        random.shuttfle(cards) #每摸一张牌洗一次牌
        return a
        
persionA = []
persionB = []
while cards:
    persionA.append(dealcards(cards))
    if cards:
        persionB.append(dealcards(cards))
````

#### 插牌

为了出牌方便，需要按照规则把牌面排好。这个游戏里为了便于找到同花顺，先按面值大小排序再按花色分组。列表自身带有`sort()`函数专门进行排序。由于这个游戏里用三元组表示扑克牌`(suit,face,figure)`,不能简单用`sort()`,需要用参数`key`指定按三元组中的哪个进行排序。

三元组和列表一样，索引从0开始，可以用索引引用三元组里的元素。下面代码中用了一种叫lambda表达式的技术。

```
persionA.sort(key=lambda x:(x[2],x[0]))   #先按x[2]排序，再按x[0]排序
persionB.sort(key=lambda x:(x[2],x[0]))
```

### 游戏循环

用户从进入游戏开始，就进入了游戏循环。在游戏循环中，玩家可以反复进行游戏操作，比如要反复多次出牌。直到用户退出游戏或者game over。哪些操作要放到游戏循环中，需要考虑游戏规则，比如一轮游戏结束后是否自动开启新一轮游戏？如果需要，洗牌、发牌的操作就要放到游戏循环中。如果不需要，一轮结束后游戏就退出了，继续玩就需要重新运行游戏。

为了简化教程，这里设定游戏是一次性的，这样洗牌、发牌的部分就不需要放到游戏循环中，但出牌的操作要反复进行多次，所以出牌的操作需要放到游戏循环中。

由于不知道要出多少次牌才能结束游戏，for循环就很难确定循环次数。这里更适合用条件循环语句while

```
# 游戏循环结构（伪码）
going = True           #循环条件，为True时执行游戏循环，为False时退出游戏循环
while going:
    do soming         #循环体部分
    if gameover:       # 如果满足游戏结束条件
        going = False   #更改循环条件为False
```

### 输入检查

前面有意忽略了很重要的一个步骤，就是检查用户输入的内容是否符合要求。用户乱输一气怎么办？如果不做处理，程序会出错。我们需要检查用户输入的只能是数字和英文逗号隔开的数字序列。

每一个需要输入的地方都要进行相应的检查。所以下面的代码要放到出牌的那段代码里。

我们自定义一个`checkinput()`函数来实现输入检查。`isdigit()`用于检查字符串是否全都是由数字构成。是返回`True`，否返回`False`。

#### 检查输入函数 checkinput()
```
# 检查用户输入只能是数字类型
# chklist: 字符串列表
# 如果chklist全都由数字字符组成，返回True,否则返回False
def checkinput(chklist):
    for x in chklist:
        if not x.isdigit():
            print("请输入数字")
            return False
    return True
```
